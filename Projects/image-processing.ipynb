```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Set matplotlib display parameters using rcParams
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['image.cmap'] = 'gray'

# Load the image
image = cv2.imread('/kaggle/input/tiger/Tiger.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
# Define a custom average filter kernel (e.g., 5x5)
kernel_size = 30
kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size ** 2)

# Apply the filter using filter2D
filtered_image = cv2.filter2D(image, -1, kernel)

# Display the original and filtered images
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.title('Filtered Image (Average)')
plt.imshow(filtered_image)

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image and resize to very small size first
image = cv2.imread('/kaggle/input/tiger/Tiger.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
small_image = cv2.resize(image_rgb, (50, 50), interpolation=cv2.INTER_AREA)

# Upscale by 8x using different methods
up_linear = cv2.resize(small_image, None, fx=8, fy=8, interpolation=cv2.INTER_LINEAR)
up_cubic = cv2.resize(small_image, None, fx=8, fy=8, interpolation=cv2.INTER_CUBIC)
up_nearest = cv2.resize(small_image, None, fx=8, fy=8, interpolation=cv2.INTER_NEAREST)

plt.figure(figsize=(15, 5))
plt.subplot(1, 5, 1)
plt.title('original image')
plt.imshow(image_rgb)
plt.axis('off')

plt.figure(figsize=(15, 5))
plt.subplot(1, 5, 2)
plt.title('small image')
plt.imshow(small_image)
plt.axis('off')

# Display results
plt.figure(figsize=(15, 5))
plt.subplot(1, 5, 3)
plt.title('INTER_LINEAR')
plt.imshow(up_linear)
plt.axis('off')

plt.subplot(1, 5, 4)
plt.title('INTER_CUBIC')
plt.imshow(up_cubic)
plt.axis('off')

plt.subplot(1, 5, 5)
plt.title('INTER_NEAREST')
plt.imshow(up_nearest)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread('/kaggle/input/erod-dil/Ero_dil.png', cv2.IMREAD_GRAYSCALE)

# Check if image loaded
if image is None:
    raise ValueError("Image not found. Check the path!")

# Create a kernel (structuring element)
kernel = np.ones((15, 15), np.uint8)

# Apply Dilation (expands white regions)
dilated = cv2.dilate(image, kernel, iterations=3)

# Apply Erosion (shrinks white regions)
eroded = cv2.erode(image, kernel, iterations=3)

# Display original, dilated, and eroded images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Original')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Dilated')
plt.imshow(dilated, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Eroded')
plt.imshow(eroded, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread('/kaggle/input/open-close/Ero_dil1.png', cv2.IMREAD_GRAYSCALE)

# Check if image loaded
if image is None:
    raise ValueError("Image not found. Check the path!")

# Create a kernel
kernel = np.ones((5, 5), np.uint8)

# Apply Opening (erosion followed by dilation)
opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

# Apply Closing (dilation followed by erosion)
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

# Display original, opening, and closing
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Original')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Opening')
plt.imshow(opening, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Closing')
plt.imshow(closing, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
